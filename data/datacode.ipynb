```python
!pip install faker
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime, timedelta
import random
```

```python
fake = Faker()
np.random.seed(42)
```

```python
dates = pd.date_range(start="2023-01-01", periods=500)

time_dim = pd.DataFrame({
    "time_id": range(1, 501),
    "date": dates,
    "day": dates.day,
    "month": dates.month,
    "quarter": dates.quarter,
    "year": dates.year
})
```

```python
region = pd.DataFrame({
    "region_id": range(1, 6),
    "region_name": ["North", "South", "East", "West", "Central"],
    "zone": ["Zone-A", "Zone-B", "Zone-C", "Zone-D", "Zone-E"]
})
```

```python
tower = pd.DataFrame({
    "tower_id": range(1, 501),
    "region": np.random.choice(region["region_name"], 500),
    "city": [fake.city() for _ in range(500)],
    "latitude": np.random.uniform(8, 37, 500),
    "longitude": np.random.uniform(68, 97, 500),
    "tower_type": np.random.choice(["Macro", "Micro", "Small Cell"], 500),
    "installation_date": [fake.date_between(start_date="-10y", end_date="-1y") for _ in range(500)]
})
```

```python
vendor = pd.DataFrame({
    "vendor_id": range(1, 6),
    "vendor_name": ["Ericsson", "Nokia", "Huawei", "Samsung", "ZTE"],
    "contract_type": ["Gold", "Silver", "Platinum", "Gold", "Silver"]
})
```

```python
equipment = pd.DataFrame({
    "equipment_id": range(1, 501),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "equipment_type": np.random.choice(["BTS", "Antenna", "Router"], 500),
    "vendor": np.random.choice(vendor["vendor_name"], 500),
    "model": [fake.lexify(text="Model-??") for _ in range(500)],
    "install_date": [fake.date_between(start_date="-5y", end_date="today") for _ in range(500)],
    "status": np.random.choice(["Active", "Faulty", "Maintenance"], 500)
})
```

```python
customer = pd.DataFrame({
    "customer_id": range(1, 501),
    "customer_type": np.random.choice(["Enterprise", "Individual"], 500),
    "service_plan": np.random.choice(["Basic", "Premium", "5G Unlimited"], 500),
    "region": np.random.choice(region["region_name"], 500)
})
```

```python
network_kpi = pd.DataFrame({
    "kpi_id": range(1, 501),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "equipment_id": np.random.choice(equipment["equipment_id"], 500),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "latency": np.random.uniform(10, 100, 500),
    "call_drop_rate": np.random.uniform(0, 5, 500),
    "throughput": np.random.uniform(20, 300, 500)
})
```

```python
alarm_event = pd.DataFrame({
    "alarm_id": range(1, 501),
    "equipment_id": np.random.choice(equipment["equipment_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "alarm_type": np.random.choice(["Power Failure", "Signal Loss", "Hardware Fault"], 500),
    "severity": np.random.choice(["Minor", "Major", "Critical"], 500),
    "event_time": [fake.date_time_this_year() for _ in range(500)],
    "resolution_time": np.random.randint(5, 300, 500)
})
```

```python
customer_complaint = pd.DataFrame({
    "complaint_id": range(1, 501),
    "customer_id": np.random.choice(customer["customer_id"], 500),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "category": np.random.choice(["Call Drop", "No Signal", "Slow Internet"], 500),
    "priority": np.random.choice(["Low", "Medium", "High"], 500),
    "resolution_time": np.random.randint(1, 48, 500)
})
```

```python
operational_cost = pd.DataFrame({
    "cost_id": range(1, 501),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "power_cost": np.random.randint(10000, 50000, 500),
    "maintenance_cost": np.random.randint(3000, 20000, 500),
    "vendor_cost": np.random.randint(2000, 15000, 500)
})
```

```python
rf_compliance = pd.DataFrame({
    "compliance_id": range(1, 501),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "rf_exposure": np.random.uniform(0.5, 3.0, 500),
    "compliance_status": np.random.choice(["Compliant", "Non-Compliant"], 500),
    "audit_date": [fake.date_this_year() for _ in range(500)]
})
```

```python
maintenance = pd.DataFrame({
    "maintenance_id": range(1, 501),
    "equipment_id": np.random.choice(equipment["equipment_id"], 500),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "maintenance_type": np.random.choice(["Preventive", "Corrective"], 500),
    "downtime_minutes": np.random.randint(10, 600, 500),
    "cost": np.random.randint(2000, 25000, 500)
})
```

```python
outage_event = pd.DataFrame({
    "outage_id": range(1, 501),
    "tower_id": np.random.choice(tower["tower_id"], 500),
    "time_id": np.random.choice(time_dim["time_id"], 500),
    "outage_reason": np.random.choice(["Power", "Fiber Cut", "Equipment Failure"], 500),
    "duration_minutes": np.random.randint(5, 720, 500),
    "affected_customers": np.random.randint(10, 5000, 500)
})
```

```python
complaint_category = pd.DataFrame({
    "category_id": range(1, 6),
    "category_name": ["Call Drop", "No Signal", "Slow Internet", "Billing Issue", "Coverage Issue"],
    "sla_hours": [4, 6, 8, 12, 10]
})
```

```python
tables = {
    "TIME_DIM": time_dim,
    "REGION": region,
    "TOWER": tower,
    "VENDOR": vendor,
    "EQUIPMENT": equipment,
    "CUSTOMER": customer,
    "NETWORK_KPI": network_kpi,
    "ALARM_EVENT": alarm_event,
    "CUSTOMER_COMPLAINT": customer_complaint,
    "OPERATIONAL_COST": operational_cost,
    "RF_COMPLIANCE": rf_compliance,
    "MAINTENANCE_ACTIVITY": maintenance,
    "OUTAGE_EVENT": outage_event,
    "COMPLAINT_CATEGORY": complaint_category
}

for name, df in tables.items():
    df.to_csv(f"{name}.csv", index=False)
```
